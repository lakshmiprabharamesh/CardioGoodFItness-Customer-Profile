## Creating Subplots to visualize boxplot()

Using Pandas we can loop through the series to create subplots.We can  visualize all our numerical variables.<br>
<br>

```python
# #Box-Whisker plot to check for outliers
#plt.subplots creates a figure with a grid of 2 rows and 3 columns
fig, ax = plt.subplots(2, 3, figsize=(20, 10))
# iterate over numerical variable to create boxplots
for variable, subplot in zip(numerical, ax.flatten()):
    sns.boxplot(data[variable], ax=subplot,color="aquamarine")
```

**Observation**:
1. The boxes shows where the outliers are mainly present.<br>  
2. The MEDIAN as the 50th percentile for each of the plots are shown except for Fitness,Education and usage. <br>
3. In Education,usage and fitness also there are outliers to consider
    Usage- 6,7 times a week are considered as outliers.
    Fitness- Fitness score of 1 i considered an outlier.
    Education - 20 and more years are considered as outliers.
